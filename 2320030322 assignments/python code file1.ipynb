# UNINFORMED SEARCH TECHINQUE BFS

```python
#BFS
graph = {
    '1': ['2', '3'],
    '2': ['1', '4', '5'],
    '3': ['1', '6', '7'],
    '4': ['2'],
    '5': ['2', '8'],
    '6': ['3'],
    '7': ['3', '9', '10'],
    '8': ['5'],
    '9': ['7'],
    '10': ['7']
}

def bfs(node):
    visited = []  # List to keep track of visited nodes.
    queue = []    # Initialize a queue.

    visited.append(node)
    queue.append(node)

    while queue:
        v = queue.pop(0)
        print(v, end=" ")

        for neigh in graph[v]:
            if neigh not in visited:
                visited.append(neigh)
                queue.append(neigh)

# Call the BFS function with a new starting node
bfs('1')
```

# Uninformed search technique DFS.

```python
graph = {
    'X': ['Y', 'Z'],
    'Y': ['X', 'W', 'V'],
    'Z': ['X', 'U'],
    'W': ['Y'],
    'V': ['Y', 'U'],
    'U': ['Z', 'V'],
}

def dfs(node, visited=None):
    if visited is None:
        visited = []
    visited.append(node)
    print(node, end="  ")
    for neigh in graph[node]:
        if neigh not in visited:
            dfs(neigh, visited)

# Call the DFS function with the new starting node 'Z'
dfs('Z')

```

# Uninformed search technique DFS **IDDFS**

```python
graph = {
    'P': ['Q', 'R'],
    'Q': ['P', 'S', 'T'],
    'R': ['P', 'U'],
    'S': ['Q'],
    'T': ['Q'],
    'U': ['R'],
}

def dfs_with_depth_limit(node, limit, visited=None):
    if visited is None:
        visited = []
    visited.append(node)
    print(node, end="  ")
    if limit <= 0:
        return
    for neigh in graph[node]:
        if neigh not in visited:
            dfs_with_depth_limit(neigh, limit - 1, visited)

def iddfs(start_node, max_depth):
    for depth in range(max_depth):
        print(f"\nDepth level {depth}: ", end="")
        dfs_with_depth_limit(start_node, depth)

# Call the IDDFS function with the new starting node 'P' and maximum depth 3
iddfs('P', 3)
```

# #A*

```python
import heapq

def astar(graph, start, goal, h):
    queue = [(h[start], start, 0)]  # Initialize with (heuristic, start node, cost so far)
    visited = set()

    while queue:
        f, node, g = heapq.heappop(queue)  # Get the node with the lowest f value
        if node in visited:
            continue
        print(node)  # Print the current node
        visited.add(node)

        if node == goal:
            break  # Stop when the goal is reached

        for neighbor, cost in graph[node]:
            if neighbor not in visited:
                new_g = g + cost  # Calculate cost to reach the neighbor
                new_f = new_g + h[neighbor]  # Total estimated cost (g + heuristic)
                heapq.heappush(queue, (new_f, neighbor, new_g))  # Push to the queue

# Define the new graph and heuristic values
graph = {
    'A': [('B', 2), ('C', 4)],
    'B': [('D', 1), ('E', 5)],
    'C': [('E', 1)],
    'D': [('F', 3)],
    'E': [('F', 2)],
    'F': []
}

h = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 4,
    'E': 3,
    'F': 0
}

# Run the A* algorithm with new inputs
astar(graph, 'A', 'F', h)
```

# BEST FOR SEARCH

```python
import heapq

def best_first_search(graph, start, goal, h):
    queue = [(h[start], start)]  # Initialize with (heuristic, start node)
    visited = set()

    while queue:
        _, node = heapq.heappop(queue)  # Get the node with the lowest heuristic value
        if node in visited:
            continue
        print(node)  # Print the current node
        visited.add(node)

        if node == goal:
            break  # Stop when the goal is reached

        for neighbor in graph[node]:
            if neighbor not in visited:
                heapq.heappush(queue, (h[neighbor], neighbor))  # Add neighbors to the queue with their heuristic value

# Define the new graph and heuristic values
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E'],
    'C': ['F'],
    'D': ['G'],
    'E': ['H'],
    'F': [],
    'G': ['H'],
    'H': []
}

h = {
    'A': 10,
    'B': 8,
    'C': 6,
    'D': 7,
    'E': 5,
    'F': 3,
    'G': 4,
    'H': 0
}

# Run the Best-First Search algorithm with new inputs
best_first_search(graph, 'A', 'H', h)
```

# #MIN MAX

```python
import math

# Define new terminal states
terminal_states = {
    'X': 10, 'Y': 7, 'Z': 5,
    'W': 3, 'V': 2, 'U': 8,
    'T': 6
}

# Define new game tree
game_tree = {
    'Root': ['X', 'Y', 'Z'],
    'X': ['W', 'V'],
    'Y': ['U', 'T'],
    'Z': [],
    'W': [],
    'V': [],
    'U': [],
    'T': []
}

def minimax(node, depth, is_maximizing):
    if node in terminal_states:
        return terminal_states[node]
    if is_maximizing:
        max_eval = -math.inf
        for child in game_tree.get(node, []):
            eval = minimax(child, depth + 1, False)
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = math.inf
        for child in game_tree.get(node, []):
            eval = minimax(child, depth + 1, True)
            min_eval = min(min_eval, eval)
        return min_eval

# Calculate the best value starting from 'Root'
best_value = minimax('Root', 0, True)

print(f"The best value for the Maximizer is: {best_value}")
```

# ALPHA-BETA

```python
import math

def alpha_beta_pruning(depth, node_index, maximizing_player, values, alpha, beta):
    if depth == 0:
        return values[node_index]
    if maximizing_player:
        max_eval = -math.inf
        for i in range(2):  # Binary tree: each node has 2 children
            eval = alpha_beta_pruning(depth - 1, node_index * 2 + i, False, values, alpha, beta)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break  # Beta cut-off
        return max_eval
    else:
        min_eval = math.inf
        for i in range(2):  # Binary tree: each node has 2 children
            eval = alpha_beta_pruning(depth - 1, node_index * 2 + i, True, values, alpha, beta)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break  # Alpha cut-off
        return min_eval

# Define new terminal node values and depth
values = [4, 2, 7, 6, 5, 8, 1, 3]
depth = 3
initial_alpha = -math.inf
initial_beta = math.inf

# Calculate the optimal value starting from the root of the game tree
optimal_value = alpha_beta_pruning(depth, 0, True, values, initial_alpha, initial_beta)
print(f"The optimal value is: {optimal_value}")
```
