# MAP COLORING

```python
class MapColoring:
    def __init__(self, graph, color_map):
        self.graph = graph
        self.num_nodes = len(graph)
        self.colors = {}
        self.color_map = color_map

    def can_color(self, node, color):
        for neighbor in self.graph[node]:
            if neighbor in self.colors and self.colors[neighbor] == color:
                return False
        return True

    def color_map_nodes(self, node):
        if node not in self.graph:
            return True
        for color in range(1, self.num_nodes + 1):
            if self.can_color(node, color):
                self.colors[node] = color
                if self.color_map_nodes(next_node(node)):
                    return True
                self.colors[node] = None
        return False

def next_node(node):
    return node + 1

if __name__ == "__main__":
    graph = {
        1: [2, 3, 4],
        2: [1, 3],
        3: [1, 2, 4],
        4: [1, 3]
    }

    color_map = {
        1: "Red",
        2: "Green",
        3: "Blue",
        4: "Yellow"
    }

    map_coloring = MapColoring(graph, color_map)
    if map_coloring.color_map_nodes(1):
        colored_nodes = {node: color_map[color] for node, color in map_coloring.colors.items()}
        print("Map coloring possible with colors:", colored_nodes)
    else:
        print("Map coloring not possible with given constraints.")

```
